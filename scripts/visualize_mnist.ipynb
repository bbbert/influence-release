```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

from __future__ import division
from __future__ import print_function
from __future__ import absolute_import
from __future__ import unicode_literals  

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

import IPython

sns.set(color_codes=True)

f = np.load('../../scr/output/mnist_small_save.npz')
train_images = np.reshape(np.array(f['train_images']),(-1,28,28))
train_labels = np.array(f['train_labels'])
validation_images = np.reshape(np.array(f['validation_images']),(-1,28,28))
validation_labels = np.array(f['validation_labels'])
test_images = np.reshape(np.array(f['test_images']),(-1,28,28))
test_labels = np.array(f['test_labels'])

def ceil_divide(numer,denom):
    return (numer-1)//denom+1
```

```python
sns.set_style('white')
fontsize=16

# 1173, 4644 good. 1891, 4936 bad. 1735, 3562 controversial
train_ids = [1173, 4644, 1891, 4936, 1735, 3562]
num_train_ids = len(train_ids)
types = ['Good', 'Good', 'Bad', 'Bad', 'Controversial', 'Controversial']
test_ids = [6558]

for j, ind in enumerate(test_ids):
    plt.figure(j)
    plt.imshow(test_images[ind])
    plt.title('Test image {} ({})'.format(ind, test_labels[ind]))
    
width = 3
height = ceil_divide(num_train_ids,width)
fig, axs = plt.subplots(height, width, sharex=True, sharey=True, figsize=(8*height, 4*width))
    
for i, row in enumerate(axs):
    for j, ax in enumerate(row):
        k = len(axs)*j+i
        if k < num_train_ids:
            ind = train_ids[k]
            ax.set_aspect('equal')
            ax.imshow(train_images[ind])
            ax.set_title('Train image {} ({}): {}'.format(ind, train_labels[ind], types[k]), fontsize=fontsize)

plt.savefig('../../scr/figs/mnist_point_visualizations.png', dpi=600, bbox_inches='tight')
```
